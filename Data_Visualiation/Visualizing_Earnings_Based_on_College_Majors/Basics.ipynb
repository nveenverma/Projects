# Visualizing Earnings Based On College Majors

- This data is taken from FiveThirtyEight (<a href="https://github.com/fivethirtyeight/data/blob/master/college-majors/recent-grads.csv">Github</a>).

- This dataset is on the job outcomes of students who graduated from college between 2010 and 2012.

### Data Dictionary
 
> <font color=blue>***Rank***</font> - Rank by median earnings (the dataset is ordered by this column).<br>
> <font color=blue>***Major_code***</font> - Major code.<br>
> <font color=blue>***Major***</font> - Major description.<br>
> <font color=blue>***Major_category***</font> - Category of major.<br>
> <font color=blue>***Total***</font> - Total number of people with major.<br>
> <font color=blue>***Sample_size***</font> - Sample size (unweighted) of full-time.<br>
> <font color=blue>***Men***</font> - Male graduates.<br>
> <font color=blue>***Women***</font> - Female graduates.<br>
> <font color=blue>***ShareWomen***</font> - Women as share of total.<br>
> <font color=blue>***Employed***</font> - Number employed.<br>
> <font color=blue>***Median***</font> - Median salary of full-time, year-round workers.<br>
> <font color=blue>***Low_wage_jobs***</font> - Number in low-wage service jobs.<br>
> <font color=blue>***Full_time***</font> - Number employed 35 hours or more.<br>
> <font color=blue>***Part_time***</font> - Number employed less than 35 hours.<br>

### <font color=Orange>***Aim***</font>
***Visualize data using Histograms, ScatterPlots, ScatterMatrix Plots and Bar Charts, and see if useful insights can be drawn from them.<br>However, the main objective is to practice the Data Visualization tools.***

## Pandas, Scatterplots

- Sample_size vs Median
- Sample_size vs unemployment_rate
- Full_time vs Median
- ShareWomen vs Unemployment_rate
- Men_vs_median
- Women_vs_median  

```python
sample_vs_median = recent_grads.plot(x="Sample_size", y="Median", kind="scatter")
sample_vs_unemployment_rate = recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind="scatter")
Full_time_vs_Median = recent_grads.plot(x="Full_time", y="Median", kind="scatter")
ShareWomen_vs_Unemployment_rate = recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter")
Men_vs_median = recent_grads.plot(x="Men", y="Median", kind="scatter")
Women_vs_median = recent_grads.plot(x="Women", y="Median", kind="scatter")
```

***From the 4th plot 'Share Women' vs 'Unemployment Rate', we can see that there is no correlation between these two points.<br>And from all other scatter plots as well, there is not much useful info which we can gather.<br> Now we will try to get some insights from Histograms***

## Pandas, Histograms

- Sample_size
- Median
- Employed
- Full_time
- ShareWomen
- Unemployment_rate
- Men
- Women

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize = (8, 16))
for x in range(4):
    ax = fig.add_subplot(4, 1, x+1)
    ax = recent_grads[cols[x]].plot(kind='hist')
    ax.set_title(cols[x])
```

```python
fig1 = plt.figure(figsize=(8, 16))
for x in range(4,8):
    ax = fig1.add_subplot(4, 1, (x-3))
    ax = recent_grads[cols[x]].plot(kind='hist')
    ax.set_title(cols[x])
```

***From the 6th histogram, we can observe that 85% of the majors have an unemployment rate less then 10%<br>Rest of the histogram dont give much useful insights, let's use Scatter Matrix Plot***

## Pandas, Scatter Matrix Plot
- Sample_size vs Median
- Sample_size vs Median vs Unemployment_rate

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[["Sample_size", "Median"]], figsize=(10, 6))
scatter_matrix(recent_grads[["Median", "ShareWomen", "Unemployment_rate"]], figsize=(15, 9))

plt.show()
```

***Here we can observe a negative correlation between ShareWomen and Median Salary, which means fields having Higher Median Salary tend to have less women ratio.<br>It is possibly due to the fact that high paying fields like engineering tend to have lesser women ratio.<br> Lets see if our theory can be showcased using Bar Plots.***

## Pandas, Grouped Bar Plots
- Women in top 10 Majors
- Women in bottom 10 Majors
- Unemployment_rate in top 10 Majors
- Unemployment_rate in bottom 10 Majors

```python
top10_Major = list(recent_grads['Major'][:10])
bottom10_Major = list(recent_grads['Major'][-10:])

top_women_unemployed = recent_grads[['ShareWomen','Unemployment_rate']][:10]
top_women_unemployed.set_index(recent_grads["Major"][:10], inplace=True)

bottom_women_unemployed = recent_grads[['ShareWomen','Unemployment_rate']][-10:]
bottom_women_unemployed.set_index(recent_grads["Major"][-10:], inplace=True)
```

```python
top_women_unemployed.plot.bar()
bottom_women_unemployed.plot.bar()
plt.show()
```

```python
fig3 = plt.figure(figsize=(16, 3))

ax = fig3.add_subplot(1, 2, 1)
ax = recent_grads['Median'][:10].plot.bar()
ax.set_xticklabels(top10_Major)

ax4 = fig3.add_subplot(1, 2, 2) 
ax4 = recent_grads['Median'][-10:].plot.bar()
ax4.set_ylim(0, 100000)
ax4.set_xticklabels(bottom10_Major)

plt.show()
```

***Here we can observe that 'Majors' with Higher Women Share have lesser Median Salary.<br>Meanwhile, 'Majors' with Lesser Women Share have higher Median Salary.***
