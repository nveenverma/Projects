### Extracting Data from CSV File into a list of lists
- We will create a function to:
    - Read data from CSV File
    - Split data into lists
    - Convert data type from string to int
    - Combine it all together into a list of lists

```python
def read_csv(csv_file):
    f = open(csv_file, "r")
    text = f.read()
    data = text.split("\n") # Splitting data on next line
    string_list = data[1:]  # Data without header
    final_list = []         # Final List in which data we will be there
    
    # Looping through data without header
    for item in string_list:
        int_fields = []      
        string_fields = item.split(",")      # Splitting data on comma
        for values in string_fields:         # Looping through each list in data
            int_value = int(values)          # Converting data type of item in list string --> integer
            int_fields.append(int_value)     # Appending the Data to temporary list
        final_list.append(int_fields)        # Appending temporary list to final list
    return final_list
#     return lst

cdc_list = read_csv("US_births_1994-2003_CDC_NCHS.csv")
cdc_list[:10]
```

### Defining function for extracting Births per Month from data

```python
def month_births(birthData):
    births_per_month = {}
    for item in birthData:
        month = item[1]
        births = item[4]
        if month in births_per_month:
            births_per_month[month] += births
        else:
            births_per_month[month] = births
    return births_per_month

cdc_month_births = month_births(cdc_list)
cdc_month_births
```

### Defining function for Extracting Births per Day from Data

```python
def dow_births(birthData):
    births_per_day = {}
    for item in birthData:
        day = item[3]
        births = item[4]
        if day in births_per_day:
            births_per_day[day] += int(births)
        else:
            births_per_day[day] = int(births)
    return births_per_day
cdc_day_births = dow_births(cdc_list)
cdc_day_births
```

### Making a common function for extracting Birth Count from Data 

```python
def calc_counts(data, column):
    no_of_births = {}
    
    for item in data:
        births = item[4]
        time_period = item[column]
        if time_period in no_of_births:
            no_of_births[time_period] = no_of_births[time_period] + births
        else:
            no_of_births[time_period] = births
    return no_of_births

cdc_year_births = calc_counts(cdc_list, 0)
cdc_month_births = calc_counts(cdc_list, 1)
cdc_dom_births = calc_counts(cdc_list, 2)
cdc_dow_births = calc_counts(cdc_list, 3)
```
